```python
from sklearn.preprocessing import MinMaxScaler , StandardScaler
from sklearn.model_selection import train_test_split
```

```python
features
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(features)
```

```python
input = scaled_data
output = scaled_data[:,0]
output
```

```python
series(input,output,length = 10,sampling_rate = 1,batch_size = 3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size = 0.2, random_state = 23, shuffle = False)
```

```python
x_train.shape, x_train.shape, y_train.shape, y_test.shape
```

```python
y_train
```

```python
train_generator = series(x_train,y_train,length = 5,sampling_rate = 1,batch_size = 3)
test_generator = series(x_test,y_test,length = 5,sampling_rate = 1,batch_size = 3)
```

```python
train_generator[0]
```

```python
import tensorflow as tf
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(128, input_shape = (5,5), return_sequences = True))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
model.add(tf.keras.layers.LSTM(128, return_sequences = True))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.LSTM(64, return_sequences = False))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
stop = tf.keras.callbacks.EarlyStopping(monitor ='val_loss', patience = 2, mode = 'min')
model.compile (loss = tf.losses.MeanSquaredError(),
               optimizer =tf.optimizers.Adam(), metrics = [tf.metrics.MeanAbsoluteError()] )

history = model.fit_generator(train_generator, validation_data = test_generator,epochs = 20,
                    shuffle = False , callbacks = [stop])
```

```python
model.evaluate_generator(test_generator)
```

```python
predict = model.predict_generator(test_generator)
```

```python
predict.shape[0]
```

```python
df_pred = pd.concat([pd.DataFrame(predict), pd.DataFrame(x_test[:,1:][5:])],axis = 1)
```

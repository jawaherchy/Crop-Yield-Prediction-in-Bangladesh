```python
import spacy
from spacy.lang.en import English
from spacy.pipeline import EntityRuler
import json
```

```python
def load_data(file):
  with open(file, "r", encoding ='utf-8') as f:
    data = json.load(f)
  return (data)
```

```python
def save_data(file,data):
  with open(file,"w", encoding = 'utf-8') as f:
    json.dump(data, f, indent =4)
```

```python
from typing import Pattern
def generate(file):
  data = load_data(file)
  print(len(data))
  new_food = []
  for item in data:
    new_food.append(item)
  for item in data:
    item = item.replace("The","").replace("the","").replace("and","").replace("And","")
    names = item.split(" ")
    for name in names:
      name = name.strip()
      new_food.append(name)
    if "(" in item:
      names = item.split("(")
      for name in names:
        name = name.replace(")","").strip()
        new_food.append(name)
    if "," in item:
      names = item.split(",")
      for name in names:
        name = name.replace("and","").strip()
        if " " in name:
          new_names = name.split()
          for x in new_names:
            x = x.strip()
            new_food.append(x)
        new_food.append(name)
  print(len(new_food))
  final_food = []
  titles = ["milk", "chocolate" "cheese", "sub","steak","fried","barbacoa","spicy","sweet","beef","chicken"]
  for food in new_food:
    if "" != food:
      final_food.append(food)
      for title in titles:
        titled_food = f"{title} {food}"
        final_food.append(titled_food)

  final_food = list(set(new_food))
  print(len(final_food))
  final_food.sort()
  return final_food
```

```python
food = generate("/content/sample_data/text.json")
```

```python
def create_training_data(file, type):
   data = generate(file)
   patterns = []
   for item in data:
     pattern = {
         "label" : type,
         "pattern" : item
     }
     patterns.append(pattern)
   return (patterns)


def generate_rules(patterns):
  nlp = English()
  ruler = nlp.add_pipe("entity_ruler")
  ruler.add_patterns(patterns)
  nlp.to_disk("food_ner")
```

```python

patterns = create_training_data("/content/sample_data/text.json", "Food")
```
